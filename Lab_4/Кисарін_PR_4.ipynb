```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Загрузити датасет, що містить інформацію про відвідуваність музеїв в різні роки.
data = pd.read_csv("museum_visitors.csv", encoding = "latin1")
data
```

```python
# 2. Вивести перші 5 рядків датасету.
print("Перші 5 рядків датасету:")
data.head()
```

```python
# 3. Провести попередній аналіз даних (визначити розмір датасету, тип даних, кількість пропусків).
print("Розмір датасету:")
print("Кількість рядків - ", data.shape[0])
print("Кількість стовбців - ", data.shape[1])
print("Типи даних датасету:", "\n", data.dtypes)
print("Кількість пропусків датасету у кожній колонці:", "\n", data.isna().sum())
```

```python
# 4. Перевірити наявність дублікатів. Якщо є, видалити їх і вивести розмір таблиці (без дублікатів).
print("Кількість дублікатів датасету:", data.duplicated().sum())
```

```python
# 5. Про які роки містить інформацію датасет.
data_years = data["Date"].str.split("-").str[0].unique()
# Вивести список років
print("Роки для яких є інформація у датасеті:")
for year in data_years:
    print(year)
```

```python
#6. Змінити назви столбців. Замінити великі букви на маленьки, пробіли на підкреслення (date, avila_adobe, firehouse_museum, chinese_american_museum, america_tropical_interpretive_center).
data.columns = ["date", 
                "avila_adobe", 
                "firehouse_museum", 
                "chinese_american_museum", 
                "america_tropical_interpretive_center"]
data.head()
```

```python
# 7. Обчислити середню кількість відвідувачів для кожного музею протягом всього періоду.
avg_visitors = round(data.drop(columns = ["date"]).mean())
print("Середні кількості відвідувачів музеїв протягом всього періоду:", "\n", avg_visitors)
```

```python
# 8. Знайти мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік
data_2018 = data[data["date"].str.startswith('2018')]

# Знайти мінімальну та максимальну кількість відвідувучів для кожного музею за 2018 рік
min_visitors_2018 = data_2018.drop(columns = ['date']).min()
max_visitors_2018 = data_2018.drop(columns = ['date']).max()

print("Мінімальна кількість відвідувачів кожного музею за 2018 рік:")
print(min_visitors_2018)
print("Максимальна кількість відвідувачів кожного музею за 2018 рік:")
print(max_visitors_2018)
```

```python
#9.	Визначити місяці з найвищою і найнижчою загальною кількістю відвідувачів серед усіх музеїв для 2015 року. Вивести назви місяців (не цифри).

import calendar

# Відфільтрувати дані для 2015 року
data_2015 = data[data["date"].str.startswith('2015')]

# Вибрати стовбці для обчислення
col_to_sum = ["avila_adobe", 
              "firehouse_museum", 
              "chinese_american_museum", 
              "america_tropical_interpretive_center"]

# Знайти місяці з найвищою та найнижчою загальною кількістю відвідувачів
total_visitors_per_month_2015 = data_2015.groupby(data_2015['date'].str.split('-').str[1].astype(int))[col_to_sum].sum().sum(axis = 1)

# Отримати назви місяців
month_names = [calendar.month_name[i] for i in total_visitors_per_month_2015.index]

month_with_highest_visitors_2015 = total_visitors_per_month_2015.idxmax()
month_with_lowest_visitors_2015 = total_visitors_per_month_2015.idxmin()

print("Місяць з найвищою загальною кількістю відвідувачів для 2015 року:", month_names[month_with_highest_visitors_2015])
print("Місяць з найнижчою загальною кількістю відвідувачів для 2015 року:", month_names[month_with_lowest_visitors_2015])
```

```python
# 10. Порівняти кількість відвідувачів музею "Avila Adobe" у літні і зимові місяці 2018 року

# Вибрати літні і зимові місяця
summer_months = [6, 7, 8] 
winter_months = [12, 1, 2]

# Вибрати дані для літніх і зимових місяців
summer = data_2018[data_2018['date'].str.split('-').str[1].astype(int).isin(summer_months)]
winter = data_2018[data_2018['date'].str.split('-').str[1].astype(int).isin(winter_months)]

# Обчислити суму відвідувачів для музею "Avila Adobe" у літні та зимові місяці
summer_visitors = summer["avila_adobe"].sum()
winter_visitors = winter["avila_adobe"].sum()

print("Кількість відвідувачів музею 'Avila Adobe' у літні місяці 2018 року:", summer_visitors)
print("Кількість відвідувачів музею 'Avila Adobe' у зимові місяці 2018 року:", winter_visitors)
```

```python
# 11. Знайти кореляцію між кількістю відвідувачів в кожному з музеїв  та датами у 2016 році
# Відфільтрувати дані для 2014 року
data_2016 = data[data['date'].str.startswith('2016')]

# Вибрати стовбці з кількістю відвідувачів у музеях
visitors_columns = ["avila_adobe", 
              "firehouse_museum", 
              "chinese_american_museum", 
              "america_tropical_interpretive_center"]
visitors_data = data_2016[visitors_columns]

# Знайти кореляцію між кількістю відвідувачів у музеях та датами
corr_matrix = visitors_data.corr()

print("Кореляція між кількістю відвідувачів у музеях та датами у 2016 році:")
print(corr_matrix)
```

Візуалізація даних

```python
# 12. Побудувати теплокарту кореляції для попереднього завдання.
plt.figure(figsize = (12, 9))
sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm")
plt.title("Графік heatmap кореляції")
plt.show()
```

```python
# 13. Побудувати графік ліній для кожного музею, показуючи їхню відвідуваність протягом 2017 року.

# Відфільтрувати дані 2017 року
data_2017 = data[data['date'].str.startswith('2017')]

# Встановити стиль seaborn
sns.set(style = "darkgrid")

# Вибрати стовбці з кількістю відвідувачів
visitors_columns = ["avila_adobe", 
              "firehouse_museum", 
              "chinese_american_museum", 
              "america_tropical_interpretive_center"]

# Побудувати графіки ліній для кожного музею
plt.figure(figsize = (15, 8))
for i in visitors_columns:
    sns.lineplot(data = data_2017, x = 'date', y = data_2017[i], label = i)
    
plt.title("Лінійні графіки відвідуваності музеїв у 2017 році")
plt.xlabel("Дата")
plt.ylabel("Кількість відвідувачів")
plt.xticks(rotation = 70)
plt.legend(loc = 'upper right')
plt.show()
```

```python
# 14. Побудувати графіки розсіювання для кожного музею за 2018 рік
# Встановити стиль seaborn
sns.set(style = "darkgrid")

# Вибрати стовбці з кількістю відвідувачів
visitors_columns = ["avila_adobe", 
              "firehouse_museum", 
              "chinese_american_museum", 
              "america_tropical_interpretive_center"]

# Побудувати графіки розсіювання для кожного музею
sns.pairplot(data = data_2018, vars = visitors_columns, height = 2)
plt.suptitle("Графіки розсіювання для відвідуваності музеїв у 2018 році")
plt.show()
```

```python
# 15. Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий.

# Встановити стиль seaborn
sns.set(style = "darkgrid")

# Вибрати стовбці з кількістю відвідувачів
visitors_columns = ["avila_adobe", 
              "firehouse_museum", 
              "chinese_american_museum", 
              "america_tropical_interpretive_center"]

# Відфільтрувати дані для кожного року (для інших років вже зроблено вище) та підрахувати сумарну відвідуваність
data_2014 = data[data['date'].str.startswith('2014')]

# Побудувати графіки гістограми з різними кольорами
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
fig.suptitle("Гістограма сумарної відвідуванності музеїв за роками")

colors = ['blue', 'green', 'yellow', "gray"]

for i, data_year in enumerate([data_2014, data_2015, data_2016, data_2017]):
    summed_visitors = data_year[visitors_columns].sum(axis = 1)
    ax = axes[i // 2, i % 2]
    sns.histplot(data = summed_visitors, bins = 30, kde = True, ax = ax, color = colors[i])
    ax.set_title(f'Рік {i + 2014}')
    ax.set_xlabel("Сумарна відвідуваність")
    ax.set_ylabel("Частота")

plt.tight_layout()
plt.show()
```
