```python
data.columns = ["customer_id", "age", "gender", "item_purchased", "category",
               "purchase_amount_usd", "location", "size", "color", "season", "review_rating",
               "subscription_status", "payment_method", "shipping_type", "discount_applied",
               "promo_code_used", "previous_purchases", "preferred_payment_method", "frequency_of_purchases"]
data
```

```python
#1.	Гістограма віку (age):
# Побудуйте гістограму віку клієнтів, використовуючи Seaborn.
sns.set_theme()
sns.histplot(data = data, x = data["age"], kde = True, element = "step")
sns.set(rc={"figure.figsize": (10, 13)})
plt.xlabel("Age of clients")
plt.ylabel("Count of clients")
plt.title("Histogram of Age")
```

```python
# 2. Порівняння розподілу покупок за статтею (item_purchased):
# Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які товари покупають клієнти найчастіше.

data_purchased = data[["item_purchased", "customer_id"]]
data_purchased_g = data_purchased.groupby(by = ["item_purchased"]).count()
data_purchased_g.reset_index(inplace = True)

plot = sns.barplot(x = data_purchased_g["item_purchased"], y = data_purchased_g["customer_id"])
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.title("Comparing purchase distribution")
plt.xlabel("Product purchased")
plt.ylabel("Count of clients")
```

```python
# 3. Залежність між віком та сумою покупок (purchase_amount):
# Використовуйте scatter plot, щоб візуалізувати залежність між віком та сумою покупок.
purchase_amount = data[["purchase_amount_usd", "age"]]
purchase_amount

age_purchase_amount = sns.scatterplot(x = purchase_amount["age"], y = purchase_amount["purchase_amount_usd"], marker = "+")
sns.set(rc={"figure.figsize": (13, 15)})
plt.title("Comparing age and purchase amount")
plt.xlabel("Age")
plt.ylabel("Count of purchases")
```

```python
# 4. Графік розподілу оцінок товарів (review_rating):
# Побудуйте графік розподілу оцінок товарів для оцінки якості.
review_rating = data[["item_purchased", "review_rating"]]
review_rating_d = review_rating.groupby(by = ["item_purchased"]).mean()
review_rating_d.reset_index(inplace = True)
review_rating_d["item_purchased"]
rrd_list = list(review_rating_d["item_purchased"])

plot = sns.barplot(x = review_rating_d["item_purchased"], y = review_rating_d["review_rating"])
plot = sns.lineplot(data = review_rating_d)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.title("Rating product distribution")
plt.xlabel("Product purchased")
plt.ylabel("Product rating")
```

```python
# 5. Категоріальна діаграма для статі (gender):
# Створіть категоріальну діаграму, щоб відобразити розподіл покупок за статтею.
data_pgender = data[["item_purchased", "customer_id", "gender"]]
data_pgender_g = data_pgender.groupby(by = ["item_purchased", "gender"]).count()
data_pgender_g.reset_index(inplace = True)
data_pgender_g
plot = sns.barplot(x = data_pgender_g["item_purchased"], y = data_pgender_g["customer_id"], hue = data_pgender_g["gender"])
plot = sns.lineplot(data = data_purchased_g)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.title("Gender purchase distribution")
plt.xlabel("Product purchased")
plt.ylabel("Count of clients")
```

```python
# 6.Залежність між рейтингом і сумою покупок (review_rating і purchase_amount):
# Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за допомогою scatter plot.
plot = sns.scatterplot(x = data["purchase_amount_usd"], y = data["review_rating"], marker = "x", color = "r")
```

```python
# 7. Сезонність покупок (season):
# Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти роблять найбільше покупок.
data_pseason = data[["customer_id", "season"]]
data_pseason_g = data_pseason.groupby(by = ["season"]).groups
data_pseason_gd = dict(data_pseason_g)
fall_count = len(data_pseason_gd["Fall"])
winter_count = len(data_pseason_gd["Winter"])
summer_count = len(data_pseason_gd["Summer"])
sprint_count = len(data_pseason_gd["Spring"])
seasons_dict = {"Fall": fall_count, "Winter": winter_count, "Summer": summer_count, "Spring": sprint_count}
seasons_data = pd.DataFrame(data = seasons_dict, index = [0])
seasons_data

plot = sns.barplot(data = seasons_data)
plt.title("Seasonality of purchases")
plt.xlabel("Season")
plt.ylabel("Count of clients")
```

```python
# 8. Графік розподілу розмірів одягу (size):
# Створіть графік розподілу розмірів одягу, щоб визначити популярність різних розмірів.
data_psize = data[["customer_id", "size"]]
data_psize_g = data_psize.groupby(by = ["size"]).groups
data_psize_gd = dict(data_psize_g)
data_psize_gd.keys()
l_count = len(data_psize_gd["L"])
m_count = len(data_psize_gd["M"])
s_count = len(data_psize_gd["S"])
xl_count = len(data_psize_gd["XL"])
size_dict = {"L": l_count, "M": m_count, "S": s_count, "XL": xl_count}
size_data = pd.DataFrame(data = size_dict, index = [0])

plot = sns.barplot(data = size_data)
plot = sns.lineplot(data = size_dict)
plt.title("Distribution popularity of size")
plt.xlabel("Sizes")
plt.ylabel("Count of clients")
```

```python
# 9. Кількість покупок відносно типу оплати (payment_method):
# Порівняйте, скільки покупок робиться за різними методами оплати за допомогою стовпчатої діаграми.
pay_met_data = data[["customer_id", "payment_method"]]
pay_met_data["customer_id"] = 1
pay_met_data
pay_met_datag = pay_met_data.groupby(by = ["payment_method"]).count()
pay_met_datag.reset_index(inplace = True)
pay_met_datag
plot = sns.barplot(x = pay_met_datag["payment_method"], y = pay_met_datag["customer_id"])
plt.title("Distribution popularity of payment methods")
plt.xlabel("Payment methods")
plt.ylabel("Count of clients")
```

```python
# 10. Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount):
# Створіть scatter plot для визначення, чи існує зв'язок між кількістю попередніх покупок та сумою поточної покупки.
prev_amount = data[["previous_purchases", "purchase_amount_usd"]]
prev_amount = sns.scatterplot(x = prev_amount["previous_purchases"], y = prev_amount["purchase_amount_usd"], color = "g", marker = "x" )
plt.title("Distribution of dependence between previous purchases and purchase amount")
plt.xlabel("Previous purchases")
plt.ylabel("Purchase amount")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn.
# Налаштуйте відображення графіка, включаючи назву графіку, підписи осей і інші необхідні параметри для кращого розуміння візуалізації.
# Додайте кольорову легенду.
def equation(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = equation(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface plot "sin(np.sqrt(x**2 + y**2), x = [-5, 5], y = [-5, 5]" visualization')
plt.show()
```
