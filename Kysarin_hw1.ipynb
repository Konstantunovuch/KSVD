```python
#9 Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
matrix_1 = np.array(np.random.randint(-5, 6, size=(3, 3)))
matrix_2 = np.array(np.random.randint(-5, 6, size=(3, 3)))
result_1 = matrix_1.dot(matrix_2)
result_2 = matrix_1 * matrix_2
print(result_1)
print(result_2)
```

```python
#10 Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
matrix = np.array(np.random.uniform(-1, 1, 9)).reshape(3, 3)
vector = np.array([matrix[0].sum(), matrix[1].sum(), matrix[2].sum()])
r_vector = vector.reshape(3, 1)
print(r_vector)
```

Робота з бібліотекою Pandas

```python
import pandas as pd
```

```python
#1 За допомогою метода read_html скачати дані з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2014)" за посиланням 
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
data = pd.read_html(url, header=0)[20]
data
```

```python
#1 Відобразити перші 10 рядків таблиці за допомогою методу head
data.head(10)
```

```python
#2 Визначити кількість рядків і стовпців в датафрейме (метод shape)
data.shape
```

```python
#3 Замініть значення "—" в таблиці значеннями NaN
data.replace("—", np.nan)
data
```

```python
#4 Визначте типи всіх стовпців за допомогою dataframe.dtypes
data.dtypes
```

```python
#5 Замініть типи нечислових стовпців числовими стовпцями. Це стовпці, у яких було розташовано символ "—"
# Всі стовбці мають числові типи, окрім регіону що є нормою.
```

```python
#6 Обчислити, скільки проміжків у кожному стовпці (використовуйте методи isnull, sum та len)
print(data.columns)
pd.isnull(data[['1950', '1960', '1970', '1990', '1992', '1994', '1996',
       '2000', '2004', '2008', '2012', '2014']]).sum()
```

```python
#7 Видалити дані для всієї країни з таблиці в останньому рядку таблиці
data_d = data.drop([27])
data_d
```

```python
#8 Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком (метод fillna)
data_d.fillna({"1950": data_d["1950"].mean(), "1960": data["1960"].mean(), "1970": data["1970"].mean(), "1990": data["1990"].mean(), 
            "1992": data["1992"].mean(), "1994": data["1994"].mean(), "1996": data["1996"].mean(), "2000": data["2000"].mean(), 
            "2004": data["2004"].mean(), "2008": data["2008"].mean(), "2012": data["2012"].mean(), "2014": data["2014"].mean()})
data_d
```

```python
#9 Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні
data_h = data_d[data_d["2014"] > data_d["2014"].mean()]
print(data_d["2014"].mean())
data_h[["регіон", "2014"]]
```

```python
#10 В якому регіоні була найвища народжуваність у 2010 та 2012 роках (окремо).
data_sort2012 = data_d.sort_values(by = "2012", ascending=False)
data_sort2012[["регіон", "2012"]]
# 2010 року у таблиці немає.
```

```python
#11 Побудуйте стовпчасту народжуваності за регіонами у 2014 році
data_plot = data_d.plot.bar(x="регіон", y="2014", rot=90, figsize=(15, 10), fontsize=18)
```
